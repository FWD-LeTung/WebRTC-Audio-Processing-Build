```python
import matplotlib.pyplot as plt
import librosa 

sr = 16000
mic = "nearend_mic_fileid_0 (1).wav"
clean = "nearend_speech_fileid_0 (1).wav"
novad = "0000_neural.wav"
webrtc = "0000_out.wav"
pyaec = "0000_pyaec.wav"
vadhead = "neural_2900_0000.wav"
sileroteacher = "0000_sileroTeacher.wav"
x, _ = librosa.load(novad,sr=sr)
x = x[160:]
y, _ = librosa.load(webrtc,sr=sr)
z, _ = librosa.load(vadhead,sr=sr)
z = z[160:]
w, _ = librosa.load(pyaec,sr=sr)
u, _ = librosa.load(mic, sr = sr)
c, _ = librosa.load(clean, sr = sr)
fig,ax = plt.subplots(nrows = 6, sharex=True, figsize=(12,12))
librosa.display.waveshow(x,sr=sr, ax=ax[0])
ax[0].set(title = 'No VAD conformer')
ax[0].label_outer()

librosa.display.waveshow(y,sr=sr, ax=ax[1], color = "aquamarine")
ax[1].set(title = 'webrtc')
ax[1].label_outer()

librosa.display.waveshow(z, sr=sr, ax=ax[2], color = "orange")
ax[2].set(title = 'vad head')
ax[2].label_outer()

librosa.display.waveshow(w, sr=sr, ax=ax[3], color = "red")
ax[3].set(title = 'Py AEC')
ax[3].label_outer()

librosa.display.waveshow(c, sr=sr, ax=ax[4], color = "mediumpurple")
ax[4].set(title = 'clean')
ax[4].label_outer()

librosa.display.waveshow(u, sr=sr, ax=ax[5], color = "green")
ax[5].set(title = 'Mic')
ax[5].label_outer()
```

```python
import soundfile as sf
import os

def cut_audio_files(
    input_files,
    output_dir,
    start_sec,
    end_sec
):
    os.makedirs(output_dir, exist_ok=True)

    for path in input_files:
        audio, sr = sf.read(path)
        
        start_sample = int(start_sec * sr)
        end_sample = int(end_sec * sr)

        cut_audio = audio[start_sample:end_sample]

        out_path = os.path.join(output_dir, os.path.basename(path))
        sf.write(out_path, cut_audio, sr)

        print(f"Saved: {out_path}, shape={cut_audio.shape}")

if __name__ == "__main__":
    input_files = [
        "0000_mic.flac",
        "0000_ref.flac",
        "000_clean.flac"
    ]

    cut_audio_files(
        input_files=input_files,
        output_dir="cut_audio2",
        start_sec=1.75,
        end_sec=8.5
    )
```
